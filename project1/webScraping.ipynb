```python
#Root Variable 
root = "https://www.football-data.co.uk/mmmz4281/"
```

```python
#creating list of leagues 
leagues = ['E0','E1','E2','E3','EC']
#Creating empty list to store populated league data
frames = []

#looping through leagues, read multiple csv and append it into a list 
for league in leagues:
    df = pd.read_csv(root + '2526' + '/' + league + '.csv')
    frames.append(df)
    
```

```python
#Creating League List
leagues = ['E0','E1','E2']
#Create Empty List to store data
frames = []

for league in leagues:
    for season in range(15,26):
        df = pd.read_csv(root + str(season) + str(season + 1) + '/' + league + '.csv')
        df.insert(1,'Season',str(season))
        frames.append(df)

```

```python
dict_countries = {
    'Spanish La Liga': 'SP1',
    'German Bundesliga': 'D1',
    'English Premier League': 'E0',
    'Scottish Premier League': 'SC0',
    'Portuguese Liga': 'P1',
    'Italian Serie A': 'I1',
    'French Ligue 1': 'F1',
}
```

```python
dict_countries['Spanish La Liga']
```

```python
dict_historical_data = {}

for league in dict_countries:
    frames = []
    for season in range(20,26):
        df = pd.read_csv(root + str(season) + str(season + 1) + '/' + dict_countries[league] + '.csv')
        df.insert(1,'Season',str(season))
        frames.append(df)
    df_concat =  pd.concat(frames)
    dict_historical_data[league] = df_concat
```
